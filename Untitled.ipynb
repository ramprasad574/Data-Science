```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
```

1.Load the data from “college.csv” that has attributes collected about private and public colleges for a particular year. We will try to predict the private/public status of the college from other attributes.

```python
College=pd.read_csv('C:/Users/rampr/Desktop/Data Science/Edureka Assignments/Module 9/Case Study 3/College.csv')
```

2.Use LabelEncoder to encode the target variable in to numerical form and split the data such that 20% of the data is set aside fortesting.

```python
College['Private']=LabelEncoder().fit_transform(College['Private'])
College.info()
```

```python
corr=College.corr()
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(corr)
```

```python
X=College
X=X.drop(labels='Private',axis=1)
Y=College[['Private']]
Y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.33,random_state=755)
```

```python
model=LinearSVC(C=1000,max_iter=1000)
```

```python
model.fit(X_train,y_train)
predicted=model.predict(X_test)
metrics.accuracy_score(y_test,predicted)
```
